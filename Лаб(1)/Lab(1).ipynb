```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('CaloriesIntensityUpload2.csv', index_col=False)

df.columns = ['id', 'Time', 'Intensity', 'Calories', 'DateHour', 'Date','TimeofDay']
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.dropna()
df.info()
```

```python
TimeofDay = pd.value_counts(df.TimeofDay)
TimeofDay = pd.DataFrame(TimeofDay)
TimeofDay.columns = ['count']
TimeofDay.plot.bar()
```

```python
TimeofDay.plot.pie(subplots=True,autopct='%.2f',figsize=(10,10))
```

```python
# Group by native country or region
df.boxplot(column='Calories',by='Date',fontsize=8,grid=False,figsize=(40,15))
```

```python
df.boxplot(column='Calories',by='TimeofDay',fontsize=8,grid=True,figsize=(15,10))
```

```python
df.boxplot(column='Calories',by='DateHour',fontsize=8,grid=True,figsize=(15,10))
```

```python
df.plot.scatter(x='Calories',y='Intensity',s=2)
```
